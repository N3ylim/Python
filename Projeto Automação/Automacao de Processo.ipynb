### Passo 1 - Importar Arquivos e Bibliotecas

```python
#Importar bibliotecas
import pandas as pd
import pprint
import win32com.client as win32
import pathlib
import os
```

```python
email = pd.read_excel("Bases de Dados/Emails.xlsx")
vendas = pd.read_excel("Bases de Dados/Vendas.xlsx")
lojas = pd.read_csv("Bases de Dados/Lojas.csv", encoding= "latin1", sep= ";")
display(email)
display(lojas)
display(vendas)
```

### Passo 2 - Definir Criar uma Tabela para cada Loja e Definir o dia do Indicador

```python
#incluir nome da loja em vendas
vendas = vendas.merge(lojas, on="ID Loja")
display(vendas)
```

```python
dicionario_lojas = {}
for loja in lojas['Loja']:
    dicionario_lojas[loja] = vendas.loc[vendas["Loja"]==loja, :]  

display(dicionario_lojas["Bourbon Shopping SP"])
```

```python
#ultima data da coluna para mandar o email
dia_indicador = vendas["Data"].max()
print(f'{dia_indicador.day}/{dia_indicador.month}')
```

### Passo 3 - Salvar a planilha na pasta de backup

```python
#identificar se a pasta já existe

lista_nomes = []
os.mkdir(r'C:\Users\Neylor Henrique\Documents\Python\Projeto Automação\Backup Arquivos Lojas')
caminho = pathlib.Path(r"Backup Arquivos Lojas")
arquivo_pasta = caminho.iterdir()

for arquivo in arquivo_pasta:
    lista_nomes.append(arquivo.name)

for loja in dicionario_lojas:
    if loja not in lista_nomes:
        nova_pasta = caminho / loja
        nova_pasta.mkdir()

#salvar dentro da pastas
    nome_arquivo = f'{dia_indicador.day}-{dia_indicador.month}-{loja}.xlsx'
    local_arquivo = caminho / loja / nome_arquivo

    dicionario_lojas[loja].to_excel(local_arquivo)


```

### Passo 7 - Criar ranking para diretoria

```python
faturamento_lojas_ano = vendas.groupby('Loja')[[ "Valor Final"]].sum()
faturamento_lojas_ano = faturamento_lojas_ano.sort_values(by='Valor Final', ascending=False)

vendas_dia = vendas.loc[vendas["Data"] == dia_indicador , :]
faturamento_lojas_dia = vendas_dia.groupby('Loja')[[ "Valor Final"]].sum()
faturamento_lojas_dia = faturamento_lojas_dia.sort_values(by='Valor Final', ascending=False)

# display(faturamento_lojas_ano)

nome_arquivo_anual = f'{dia_indicador.year} - Ranking Anual.xlsx'
faturamento_lojas_ano.to_excel(r'Backup Arquivos Lojas\{}'.format(nome_arquivo_anual))

# display(faturamento_lojas_dia)

nome_arquivo_diário = f'{dia_indicador.day}-{dia_indicador.month} - Ranking Diário.xlsx'
faturamento_lojas_dia.to_excel(r'Backup Arquivos Lojas\{}'.format(nome_arquivo_diário))
```
