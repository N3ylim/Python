```python
import pygame
import random
import sys

# Inicialização do Pygame
pygame.init()

# Definindo as dimensões da janela do jogo
largura = 1200
altura = 720
janela = pygame.display.set_mode((largura, altura))
pygame.display.set_caption("Jogo de Corrida")

# Definindo as cores
preto = (0, 0, 0)
branco = (0, 0, 0)
vermelho = (148, 0, 211)

# Definindo o tamanho do carro
carro_largura = 30
carro_altura = 31

# Carregando a imagem do carro
carro_imagem = pygame.image.load('carro.png')

def colisao():
    # Encerrar o jogo após a colisão
    pygame.quit()
    sys.exit()

# Função para exibir o carro na janela
def desenhar_carro(x, y):
    janela.blit(carro_imagem, (x, y))

# Função principal do jogo
def jogo():
    # Posição inicial do carro
    carro_x = largura * 0.45
    carro_y = altura * 0.8

    # Variáveis de movimento do carro
    movimento_esquerda = False
    movimento_direita = False

    # Lista de obstáculos
    obstaculos = []

    # Velocidade dos obstáculos
    obstaculo_velocidade = 10

    # Controle de velocidade do jogo
    clock = pygame.time.Clock()

    # Loop principal do jogo
    while True:
        # Eventos do jogo
        for evento in pygame.event.get():
            if evento.type == pygame.QUIT:
                pygame.quit()
                sys.exit()
            if evento.type == pygame.KEYDOWN:
                if evento.key == pygame.K_LEFT:
                    movimento_esquerda = True
                if evento.key == pygame.K_RIGHT:
                    movimento_direita = True
            if evento.type == pygame.KEYUP:
                if evento.key == pygame.K_LEFT:
                    movimento_esquerda = False
                if evento.key == pygame.K_RIGHT:
                    movimento_direita = False

        # Atualização da posição do carro
        if movimento_esquerda:
            carro_x -= 7
        if movimento_direita:
            carro_x += 7

        # Atualização da posição dos obstáculos
        for obstaculo in obstaculos:
            obstaculo['y'] += obstaculo_velocidade

        # Criação de novos obstáculos
        if len(obstaculos) < 10:  # Defina o número desejado de obstáculos simultâneos
            obstaculo_x = random.randrange(0, largura - carro_largura)
            obstaculo_y = random.randrange(-200, -100)
            obstaculos.append({'x': obstaculo_x, 'y': obstaculo_y})

        # Limpeza da janela
        janela.fill(branco)

        # Desenho dos obstáculos
        for obstaculo in obstaculos:
            pygame.draw.rect
        # Desenho dos obstáculos
        for obstaculo in obstaculos:
            pygame.draw.rect(janela, vermelho, [obstaculo['x'], obstaculo['y'], carro_largura, carro_altura])

        # Desenho do carro
        desenhar_carro(carro_x, carro_y)

        # Verificação de colisão entre o carro e os obstáculos
        for obstaculo in obstaculos:
            if carro_y < obstaculo['y'] + carro_altura:
                if carro_x > obstaculo['x'] and carro_x < obstaculo['x'] + carro_largura or carro_x + carro_largura > obstaculo['x'] and carro_x + carro_largura < obstaculo['x'] + carro_largura:
                    colisao()

        # Remoção dos obstáculos que saem da tela
        obstaculos = [obstaculo for obstaculo in obstaculos if obstaculo['y'] < altura]

        # Atualização da janela
        pygame.display.update()

        # Controlando a velocidade do jogo
        clock.tick(60)  # Definindo 60 FPS (quadros por segundo)

# Chamada da função principal do jogo
jogo()
```
