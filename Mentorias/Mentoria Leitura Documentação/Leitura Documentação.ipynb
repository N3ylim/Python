```python
# pandas
import pandas as pd

df = pd.DataFrame(
    {
        "Name": [
            "Braund, Mr. Owen Harris",
            "Allen, Mr. William Henry",
            "Bonnell, Miss. Elizabeth",
        ],
        "Age": [22, 35, 58],
        "Sex": ["male", "male", "female"],
    }
)

display(df)
display(df['Age'])
df["Age"].max()
# pd.DataFrame
# pd.DataFrame => recebe um dicionario como paramentro e cria a tabela a partir do dicionario
```

```python
ages = pd.Series([22, 35, 58], name="Age")

display(ages)
ages.max()
```

#### Funções Específicas

```python
# datas no pandas
import datetime as dt

tabela = pd.read_excel('Contratos.xlsx')
tabela['Data Pagamento'] = pd.to_datetime(tabela['Data Pagamento'])
tabela['Data Pagamento 2'] = pd.to_datetime(tabela['Data Pagamento 2'], format='%m-%d-%y')
print(tabela.dtypes)
display(tabela)

```

```python
# funcao merge do pandas

email = pd.read_excel('Emails.xlsx')

tabela_nova = tabela.merge(email, left_on='Cliente', right_on='Nome')
tabela_nova = tabela_nova.drop(columns='Nome')

display(email)
display(tabela)
display(tabela_nova)

```
