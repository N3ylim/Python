```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

servico = Service(ChromeDriverManager().install())
navegador = webdriver.Chrome(service=servico)
```

```python
import os

caminho = os.getcwd()
arquivo = caminho + r"\Sites\Pagina Hashtag.html"
navegador.get(arquivo)
```

#### Botão que abre outra janela

```python
navegador.find_element(By.CLASS_NAME, 'wp-image-13720').click()
```

```python
aba_original = navegador.window_handles[0]
nova_aba = navegador.window_handles[1]
navegador.switch_to.window(nova_aba)
```

```python
navegador.find_element(By.ID, 'firstname').send_keys('Ney')
navegador.find_element(By.ID, 'email').send_keys('Ney@gmail.com')
navegador.find_element(By.ID, 'phone').send_keys('111223344556')
```

```python
navegador.switch_to.window(aba_original)
navegador.find_element(By.CLASS_NAME, 'wp-image-13719').click()
```

#### Ver os titulos de todas as abas

```python
for aba in navegador.window_handles:
    navegador.switch_to.window(aba)
    print(navegador.title)
```

```python
navegador.quit()
```
