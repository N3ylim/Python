### Vamos analisar o IBOV

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plot
import numpy as np

data_inicio = "2023-01-01"
data_fim = "2023-06-20"
simbolo = "^BVSP"  # Índice Bovespa

# Buscar os dados
dados = yf.download(simbolo, start=data_inicio, end=data_fim)

# Exibir os dados
display(dados)

```

### Analisando o Gráfico

```python
dados["Adj Close"].plot(figsize = (13,5))
```

### Retorno do IBOV

```python
retorno = dados['Adj Close'][-1] / dados['Adj Close'][0] - 1
print(f"Retorno IBOV: {retorno:.2%}")
```

### Analisando com Média Móvel

```python
dados["Adj Close"].plot(figsize = (15,5), label="IBOV")
dados["Adj Close"].rolling(21).mean().plot(label="MM21")
dados["Adj Close"].rolling(34).mean().plot(label="MM34")

plot.legend()
plot.show()
```
