#### Pegando cotações semanais - CSV

```python
import pandas as pd
from io import StringIO

##Diario
# replace the "demo" apikey below with your own key from https://www.alphavantage.co/support/#api-key
url = f'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=NVDA&apikey={chave_api}&datatype=csv'
r = requests.get(url)

# pprint.pprint(r.text)
tabela = pd.read_csv(StringIO(r.text))
display(tabela)
```

### Cotações

```python
acoes = ['ITUB4', 'ABEV3', 'BBAS3']

compilada = pd.DataFrame()

for acao in acoes:
    # replace the "demo" apikey below with your own key from https://www.alphavantage.co/support/#api-key
    url = f'https://www.alphavantage.co/query?function=GLOBAL_QUOTE&symbol={acao}.SAO&apikey={chave_api}&datatype=csv'
    r = requests.get(url)

    # pprint.pprint(r.text)
    tabela = pd.read_csv(StringIO(r.text))
    lista_tabela = [compilada, tabela]
    compilada = pd.concat(lista_tabela)
display(compilada)
```

### Encontrar ativos

```python
url = f'https://www.alphavantage.co/query?function=SYMBOL_SEARCH&&keywords=samsung&apikey={chave_api}&datatype=csv'
r = requests.get(url)

# pprint.pprint(r.text)
tabela = pd.read_csv(StringIO(r.text))
display(tabela)
```
