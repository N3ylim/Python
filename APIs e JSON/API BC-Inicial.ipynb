### API do Banco Central

https://dadosabertos.bcb.gov.br/dataset?res_format=API

```python
import requests
import pprint

link = "https://olinda.bcb.gov.br/olinda/servico/mecir_dinheiro_em_circulacao/versao/v1/odata/informacoes_diarias?$top=1000&$orderby=Quantidade%20desc&$format=json"

requiscao = requests.get(link)
info = requiscao.json()
# pprint.pprint(info)
```

```python
import pandas as pd

tabela = pd.DataFrame(info["value"])
tabela["Quantidade"] = tabela["Quantidade"].map("R${:,.2f}".format)


display(tabela)
```

```python
#Pegar todas as informações com várias requisições 

tabela_final = pd.DataFrame()
skip = 0

while True:
    link = f'https://olinda.bcb.gov.br/olinda/servico/mecir_dinheiro_em_circulacao/versao/v1/odata/informacoes_diarias?$top=10000&$skip={skip}&$orderby=Data%20desc&$format=json'
    requiscao = requests.get(link)
    info = requiscao.json()
    tabela = pd.DataFrame(info["value"])
    if len(info['value']) < 1:
        break
    tabela_final = pd.concat([tabela_final, tabela])
    skip += 10000

display(tabela_final)
```
